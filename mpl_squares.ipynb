# Plot a Simple Line Graph


```python
import matplotlib.pyplot as plt

squares = [1,4,9,16,25]

fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

# Adding Labels and Changing Line Thickness

```python
import matplotlib.pyplot as plt

squares = [1,4,9,16,25]

fig, ax = plt.subplots()

ax.plot(squares, linewidth=3)
#Add chart labels and chart title
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

#set size of thick labels 
ax.tick_params(axis= 'both', labelsize = 14)



plt.show()
```

# Correcting the Plot 

```python
import matplotlib.pyplot as plt

input_values =[1,2,3,4,5]
squares = [1,4,9,16,25]

fig, ax = plt.subplots()

ax.plot(input_values,squares, linewidth=3)
#Add chart labels and chart title
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

#set size of thick labels 
ax.tick_params(axis= 'both', labelsize = 14)



plt.show()
```


# Using Built-in Styles

```python
plt.style.available
```

```python

plt.style.use('seaborn-dark')

fig, ax = plt.subplots()
ax.plot(input_values, squares)

ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize =14)
ax.set_title("Square Numbers", fontsize = 24)

ax.tick_params(axis = 'both', labelsize = 14)

plt.show
```


# Styling and plotting Individual Points with Scatter()

```python
plt.style.use('seaborn-dark')

fig, ax = plt.subplots()
ax.scatter(input_values, squares)
ax.set_title("Square Numbers", fontsize = 25)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Sqaure Values", fontsize = 14)

ax.tick_params(axis='both', labelsize = 14)

plt.show()
```
