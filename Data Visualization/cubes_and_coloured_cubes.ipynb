15-1. Cubes: A number raised to the third power is a cube. Plot the first five
cubic numbers, and then plot the first 5000 cubic numbers.
15-2. Colored Cubes: Apply a colormap to your cubes plot.

```python
import matplotlib.pyplot as plt

x_values = range(1, 6)
y_values = [x**3 for x in x_values]

fig , ax = plt.subplots()

plt.style.use('seaborn')

ax.set_title("Cube Numbers", fontsize = 25)
ax.set_ylabel("Values Cubed", fontsize = 14)
ax.set_xlabel("Values", fontsize =14)
ax.tick_params(axis = 'both', labelsize = 14)

ax.axis([0,6,0,130])

ax.scatter(x_values,y_values,color="blue",s=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]


fig , ax = plt.subplots()

plt.style.use('seaborn')

ax.set_title("Cube Numbers", fontsize = 25)
ax.set_ylabel("Values Cubed", fontsize = 14)
ax.set_xlabel("Values", fontsize =14)
ax.tick_params(axis = 'both', labelsize = 14)

ax.axis([0,5500,0,126000000000])

ax.scatter(x_values,y_values,color="blue",s=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]


fig , ax = plt.subplots()

plt.style.use('seaborn')

ax.set_title("Cube Numbers", fontsize = 25)
ax.set_ylabel("Values Cubed", fontsize = 14)
ax.set_xlabel("Values", fontsize =14)
ax.tick_params(axis = 'both', labelsize = 14)

ax.axis([0,5500,0,126000000000])

ax.scatter(x_values,y_values,c = y_values ,cmap= plt.cm.Blues,s=10)
plt.show()
```
